```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
def profile_plot(file_path, title, fig_name, x_inv = False, y_inv = True, save = False):
    """
    Plots the profile created by the C++ simulation
    title : title of the plot
    x_inv : inverts the x axis, by default set to False
    y_inv : inverts the y axis, by default set to True
    save : if True saves the plot to a png file with fig_name as name
    """
    # Load the CSV file
    grid_data = pd.read_csv(file_path)
    
    # Define a mapping from cell types to numeric values
    cell_type_mapping = {
        'VACUUM': 0,
        'MASK': 1,
        'SI': 2,
        'SIF': 3,
        'SIF2': 4,
        'SIF3': 5,
        'SIF4': 6
    }
    
    # Apply the mapping to the DataFrame
    numeric_grid = grid_data.replace(cell_type_mapping)
    
    # Convert the DataFrame to a numpy array for plotting
    grid_array = numeric_grid.values
    
    # Plotting the heatmap
    plt.figure(figsize=(20, 10))
    sns.heatmap(grid_array, cmap='viridis', cbar_kws={'label': 'Cell Type'})
    plt.title(title)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    if (x_inv):
        plt.gca().invert_xaxis()
    if (y_inv):
        plt.gca().invert_yaxis()
    if (save):
        plt.savefig("figures/" + fig_name + ".png")
    plt.show()
```

```python
profile_plot("data/testing_straight.csv",
             "zoom_straight asy = 0.3, p = 80000, 0.3 : 0.6",
             "test_zoom_straight_asy03_p80000_v_03_06",
             save=False)
```
